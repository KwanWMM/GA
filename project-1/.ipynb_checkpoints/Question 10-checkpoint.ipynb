```python
# Code to read in pokedex info
raw_pd = ''
pokedex_file = 'pokedex_basic.csv'
with open(pokedex_file, 'r') as f:
    raw_pd = f.read()
    
# the pokedex string is assigned to the raw_pd variable
```

```python
clean_list = [[float(index)if (index >0 and index_2==0) else float(raw_pd.splitlines()[index].split(',')[index_2].replace('"',"")) if raw_pd.splitlines()[index].split(',')[index_2].replace('"',"")[0] in '0123456789' else float(-1) if raw_pd.splitlines()[index].split(',')[index_2].replace('"',"") == '' else raw_pd.splitlines()[index].split(',')[index_2].replace('"',"") for index_2 in range(len(raw_pd.splitlines()[index].split(',')))] for index in range(len(raw_pd.splitlines()))]
```

```python
clean_list
```

```python
full_pokedex = {}

def full_pokedex_generator(pokemon_table):
    #iterate through every row except the header
    for row_num in range(1,len(pokemon_table)):
        #Add the pokemon into the dictionary using the pokemon id as the key
        full_pokedex[pokemon_table[row_num][0]]={
                                                    #the values are a dictionary of the pokemon's stats
                                                    'Name':pokemon_table[row_num][1],
                                                    'Type':pokemon_table[row_num][2],
                                                    'Total':pokemon_table[row_num][3],
                                                    'HP':pokemon_table[row_num][4],
                                                    'Attack':pokemon_table[row_num][5],
                                                    'Defense':pokemon_table[row_num][6],
                                                    'SpecialAttack':pokemon_table[row_num][7],
                                                    'SpecialDefense':pokemon_table[row_num][8],
                                                    'Speed':pokemon_table[row_num][9]
                                                }
    return full_pokedex
```

```python
full_pokedex_generator(clean_list)
```
