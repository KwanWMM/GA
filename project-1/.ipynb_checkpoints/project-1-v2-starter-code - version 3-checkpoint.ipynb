## 1. Defining a player

---

The player variables are:

    player_id : id code unique to each player (integer)
    player_name : entered name of the player (string)
    time_played : number of time played the game in minutes (float)
    player_pokemon: the player's captured pokemon (dictionary)
    gyms_visited: ids of the gyms that a player has visited (list)
    
Create the components for a player object by defining each of these variables. The dictionary and list variables should just be defined as empty; you can use any (correctly typed) values for the others.

````python
#Create a player class
class Player():
    def __init__(self,player_id,player_name,time_played):
        player_pokemon = {}
        gyms_visited = []
        
        #create the class attributes
        self.player_id = player_id
        self.player_name = player_name
        self.time_played = time_played
        self.player_pokemon = player_pokemon
        self.gyms_visited = gyms_visited
    
    #Include a method for extending gyms visited (so multiple gyms can be included at a time)
    def add_gym(self,gym_name):
        self.gyms_visited.extend(gym_name)

#Cerate Player 1, ash
ash = Player(1,'Ash',25.0)
````

## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def player_power(p_dict, mon_dex, p_id):
    #Variable for the player's power
    p_power = 0
    
    #Iterate through each pokemon that the player has 
    for mon_id in p_dict[p_id]['player_pokemon']:
        #Add the pokemon's attack and defence stats to the player's power
        p_power += (p_dict[p_id]['player_pokemon'][mon_id]['attack']+p_dict[p_id]['player_pokemon'][mon_id]['defence'])
        
    #Print the notification and return player's power    
    print("{name}'s power is {power}.".format(name = p_dict[p_id]['player_name'],power = p_power))
    return p_power
````

---
Your function should:

1. Take a dictionary as a parameter with keys matching the features of the Pokedex, filtering by exact match for string type values, and/or filter continuous variables specified value that is greater than or equal to the dictionary key parameter.
1. Return multiple elements from the Pokedex

Example:

```python

# Only filter based on parameters passed
filter_options = {
    'Attack':   25,

}

# Return records with attack >= 24, defense >= 30, and type == "Electric"

# Example output:
# [{'Attack': 30.0,
#  'Defense': 50.0,
#  'HP': 40.0,
#  'Name': 'Voltorb',
#  'SpecialAttack': 55.0,

#  'Speed': 100.0,
#  'Total': 330.0,
#  'Type': 'Electric'},
#  ... etc
#  ]

```



````python
def filtered_pokedex(pokedex,filter_dict):
    #Create the list to be returned at the end
    #Pokemon that mean filter conditions will be added to the list
    filtered_list = []
    
    #Iterate through every pokemon to test against filter conditions
    for pokemon in pokedex:
        #Create a counter for the number of criteria the pokemon meet
        counter = 0
        #Create the 
        pass_counter = len(filter_dict)
        
        for criteria in filter_dict:
            if criteria=='Type' or criteria=='Name':
                if pokedex[pokemon][criteria] == filter_dict[criteria]:
                    filtered_list.append(pokedex[pokemon])
                    break
                else:
                    pass_counter -= 1
            elif pokedex[pokemon][criteria] >= filter_dict[criteria]:
                counter += 1
        if counter == pass_counter and pass_counter !=0:
            filtered_list.append(pokedex[pokemon])
    
    return filtered_list
   
````

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?

### 9.2

The game is no fun if the characters are wildly unbalanced! Are any characters "overpowered", which we'll define as having a "Total" more than three standard deviations from the population mean?

````python
#Function to check if any pokemon is overpowered.
def who_is_overpowered():
    #Iterate through every pokemon in the pokedex dictionary.
    for key in full_pokedex:
        #Check if the pokemon's 'Total' attribute is greater than 3 standard deviations from the mean.
        if full_pokedex[key]['Total'] > (total_mean+3*total_std):
            #if pokemon is overpowered, announce it.
            print(f"{full_pokedex[key]['Name']} is overpowered")
````

## 10. Calibrate the frequency of Pokemon

The design team wants you to make the powerful Pokemon rare, and the weaklings more common. How would you set the probability $p_i$ of finding Pokemon *i* each time a player visits a gym?

Write a function that takes in a Pokedex number and returns a value $p_i$ for that character.

Hint: there are many ways you could do this. What do _you_ think makes sense? Start with simplifying assumptions: for example, you could assume that the probabilities of encountering any two Pokemon on one visit to a gym are independent of each other.

**Answer: **

Base assumptions: 
probabilities of encountering any two Pokemon on one visit to a gym are independent of each other.

Each time you visit a gym, you will meet a pokemon, P(S) = 1  
Sample space: {1,2,3,4,...,800}  
Meeting pokemon are disjoint events, you cannot meet two pokemon at once.  
Therefore, you will meet one of the 800 pokemon, summation of P(_id1_) to P(_id800_) = 1.

Assumption 2: Strongest pokemon should be the hardest to catch.

Assumption 3: Strength of pokemon will be determined via the 'Total' stat.

Assumption 3: Players should be motivated to continue playing by not constantly catching the weakest pokemon, the weakest pokemon should not be the easiest to catch otherwise the player is spammed by only weak pokemon.
Therefore, probability should not a follow linear line based on pokemon's Total stat.

Assumption 4: It should be harder to catch the strongest pokemon than the weakest pokemon.
Therefore, the probability distribution should have a positive skew. 

By sorting the pokemon in order of strength, then mapping the list onto a distribution with a positive skew, 
the probability of the pokemon can then be decided.

**I was unable to find a formula to create a distribution with a normal skew, so I decided to use sigmoid curves to generate a graph that looks somewhat similar.**

````python
#import neccesary libraries
import math
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
#Create a function for generating sigmoid values to generate sigmoid curves
def sigmoid(num_list):
    sigmoid_list=[]
    
    #Iterate through the list entered into the function
    for item in num_list:
        
        #Append the sigmoid values to sigmoid_list
        sigmoid_list.append(1/(1+math.exp(-item)))
    
    #return the list of sigmoid values
    return sigmoid_list
````

````python
#Create lists for both the left and right of the sigmoid curve 
#There are 1000 elements in total as the first 200 will be truncated to create the positive skew
left = [num/50 for num in range(-250,250)]
right = [num/50 for num in range(-250,250)]
````

````python
#create a function to calculate the probability
def probability(num):
    #The probability is simply the value divided by the sum of all possible values
    #Do not count the first 200 values of sigmoid(left) as they will be truncated to create the positive skew
    return (num/(sum(sigmoid(left)[200:])+sum(sigmoid(right))))

#Create the list of probabilities for the left hand side of the curve (first 200 truncated to create the positive skew)
prob_left = [probability(num) for num in sigmoid(left)[200:]]

#Create the list of probabilities for the right hand side of the curve
#sigmoid(right) is reversed as it is a mirror image of sigmoid(left))
prob_right = [probability(num) for num in sigmoid(right)[::-1]]

#list of total probabilities
prob_combined = prob_left+prob_right
````

````python
#Generate the x-axis, 800 numbers representing the pokemon sorted by strength
x_axis = [num for num in range(800)]

#Plot the Pokeon ID aginst the porbability of catching them
plt.xlabel('Pokemon ID')
plt.ylabel('Probability')
plt.plot(x_axis,prob_combined)
````

**Now to order the pokemon in order of strength ('Total' stat)**

````python
for_sorting=[]

#Iterate through the pokedex and create a list with some pokemon stats
for pokemon in full_pokedex:
    #Values to be used for sorting are firstly Total, followed by Attack then Defense and eventually Name in the event of ties.
    for_sorting.append([full_pokedex[pokemon]['Total'],full_pokedex[pokemon]['Attack'],full_pokedex[pokemon]['Defense'],full_pokedex[pokemon]['Name']])
                        
#Sort the pokemon from weakest to strongest
sorted_names = [pokemon[3] for pokemon in sorted(for_sorting)]
````

````python
#Create an empty dictionary that will be holding the pokemon names and their probability of appearing as key:value pairs
probability_dictionary={}

#Iterate through the sorted list of pokemon 
for pokemon in sorted_names:
    #Add the pokemon's Pokedex number and probability of appearing into probability_dictionary
    probability_dictionary[pokemon]=prob_combined[sorted_names.index(pokemon)]
````

````python
#Create a function that takes in a pokemon id number 
def pokemon_probability(num):
    #return the probability by checking the pokemon's name against probability_dictionary
    return(probability_dictionary[full_pokedex[num]['Name']])
````
