```python
#Check what the different one is
basement_height[~basement_height.isnull()]
```

```python
#Assume the different one is an error and replace with None.
#Convert the null BsmtQual to None and BsmtExposure to NoBasement.
house['BsmtQual']=house['BsmtQual'].fillna('None')
house['BsmtExposure']=house['BsmtExposure'].fillna('NoBasement')
```
