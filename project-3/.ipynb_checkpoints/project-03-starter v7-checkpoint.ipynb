```python
#Assume that the LotFrontages are similar within the same neighbourhood
#Create functions for impute mean and median for LotFrontage based on the neighbourhood it is found in 
def impute_frontage_mean(row):
    lotfrontage = row[0]
    neighborhood = row[1]
    
    if pd.isnull(row[0]):
        return np.mean(house[house['Neighborhood']==row[1]]['LotFrontage'].dropna())
    else:
        return row[0]
    
def impute_frontage_median(row):
    lotfrontage = row[0]
    neighborhood = row[1]
    
    if pd.isnull(row[0]):
        return np.median(house[house['Neighborhood']==row[1]]['LotFrontage'].dropna())
    else:
        return row[0]
```

```python
def get_dummies(df): 
    '''
    DOCSTRING: This function takes in a DataFrame and converts all columns into dummy variables. If the column has null values, the null values will be dropped, making that the default for hot-one encoding. If the column has no null values, the element that occurs the most will be regarded as the default for hot-one encoding and dropped.
    INPUT: DataFrame
    OUTPUT: DataFrame
    '''
    #Create an empty DataFrame to append to
    dummy_cat_col = pd.DataFrame()

    #iterate through each column in df
    for col in df.columns:
        #if the col has no null values
        if df[col].isnull().sum() == 0:
            #create dummy variables from the column
            new_col = pd.get_dummies(df[col])
            #drop the column that was the most common occurring element in the original column (becomes the default for hot-one encoding)
            new_col.drop(df[col].value_counts().first_valid_index(),axis=1,inplace=True)
            #rename the columns with original column name included
            new_col.columns=[col+'_'+ str(name) for name in new_col.columns]
            #append to the list of dummy variables 
            dummy_cat_col = pd.concat([dummy_cat_col,new_col],axis=1)
        else:
            #create dummy variables from the column
            new_col = pd.get_dummies(df[col])
            #rename the columns with original column name included
            new_col.columns=[col+'_' + str(name) for name in new_col.columns]
            #append to the list of dummy variables 
            dummy_cat_col = pd.concat([dummy_cat_col,new_col],axis=1)
    return dummy_cat_col
```

Next, I'll try regularization methods to deal with the possible multicollinearity and reduce overfitting. 

```python
def regularization_tests(X,y,cv=3):
    '''
    DOCSTRING This function runs RidgeCV, LassoCV and ElasticNetCV to find the optimal alpha(s) and L1 ratio, before performing cross-validation testing and returning the selected models and a dataframe with the mean R2 scores and standard deviations for the regularizations.
    INPUT DataFrame of independent variables (X), array/series of dependent variable (y), number of cross validations to run
    OUTPUT: Selected ridge model, lasso model, elastic net model and scores(mean R2 score, std) DataFrame
    '''
    mean_scores=[]
    stds = []

    #Compute Ridge path with coordinate descent to obtain the optimal alpha 
    ridge_cv = RidgeCV(alphas=np.logspace(-3,5,200), cv=cv)
    ridge_cv.fit(X,y)
    #Run a Ridge Regression using the optimal alpha calculated above
    ridge_reg = Ridge(alpha=ridge_cv.alpha_)
    scores = cross_val_score(ridge_reg,X,y, cv=cv)
    #Append the mean score and std to the respective lists 
    mean_scores.append(np.mean(scores))
    stds.append(np.std(scores))
    
    #Compute Lasso path with coordinate descent to obtain the optimal alpha 
    lasso_cv = LassoCV(n_alphas=200,cv=cv)
    lasso_cv.fit(X,y)
    #Run a Lasso Regression using the optimal alpha calculated above
    lasso_reg = Lasso(alpha=lasso_cv.alpha_)
    scores = cross_val_score(lasso_reg,X,y, cv=3)
    #Append the mean score and std to the respective lists 
    mean_scores.append(np.mean(scores))
    stds.append(np.std(scores))
    
    #Compute elastic net path with coordinate descent to obtain the optimal alpha and l1 ratio
    enet_cv = ElasticNetCV(l1_ratio=np.linspace(0.1,1,100),n_alphas=200, cv=cv)
    enet_cv.fit(X,y)
    
    #Run an Elastic Net Regression using the optimal alpha calculated above
    enet_reg = ElasticNet(alpha=enet_cv.alpha_,l1_ratio=enet_cv.l1_ratio_)
    scores = cross_val_score(enet_reg,X,y, cv=3)
    #Append the mean score and std to the respective lists 
    mean_scores.append(np.mean(scores))
    stds.append(np.std(scores))
    
    score_df = pd.DataFrame({'Regularisation':['Ridge','Lasso','Elastic Net'],'Mean R2 scores':mean_scores,'Standard Deviation':stds})
    return (ridge_reg, lasso_reg, enet_reg, score_df)
    
```

```python
ridge_model, lasso_model, elastic_model, scores_table = regularization_tests(Xs_train,y_train)
```

```python
scores_table
```

```python
def training_test_results(ridge, lasso, elastic, X_train, y_train, X_test, y_test):
    '''
    DOCSTRING: This function takes in  Ridge, Lasso and Elastic Net Model as well as the standardised X training DataFrame, y training targets, standardised X test DataFrame, y test targets
    INPUT: Ridge model, Lasso model, Elastic Net model, standardised X training data, y training target, standardised X test data, y test target
    OUTPUT: scores (R2 score,mse) DataFrame
    '''
    
    scores=[]
    mse=[]
    
    ridge.fit(X_train,y_train)
    scores.append(ridge.score(X_test,y_test))
    mse.append(mean_squared_error(y_test, ridge.predict(X_test)))

    lasso.fit(X_train,y_train)
    scores.append(lasso.score(X_test,y_test))
    mse.append(mean_squared_error(y_test, lasso.predict(X_test)))
    
    elastic.fit(X_train,y_train)
    scores.append(elastic.score(X_test,y_test))
    mse.append(mean_squared_error(y_test, elastic.predict(X_test)))
    
    results=pd.DataFrame({'Regularisation':['Ridge','Lasso','Elastic Net'],'Scores':scores,'MSE':mse})
    
    return results
```

```python
training_test_results(ridge_model, lasso_model, elastic_model, Xs_train, y_train, Xs_test,y_test)
```
